```python
def converter_em_lista(entrada):
    return list(map(int, entrada.split()))

def verificar_ordem_crescente(lista):
    return lista == sorted(lista)

def verificar_ordem_decrescente(lista):
    return lista == sorted(lista, reverse=True)

# Entrada do usuário
entrada = input()
lista = converter_em_lista(entrada)
print("Antes:", lista)

if verificar_ordem_crescente(lista):
    print("A lista está ordenada em ordem crescente.")
elif verificar_ordem_decrescente(lista):
    print("A lista está ordenada em ordem decrescente.")
else:
    print("A lista não está ordenada nem em ordem crescente, nem em ordem decrescente.")
```
