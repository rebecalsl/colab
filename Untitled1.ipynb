```python
def fatorial(n, nivel=0):
    print("Calculando fatorial de", n)
    if n == 0 or n == 1:
        print(f"Fatorial de {n} é 1 (caso base)")
        return 1
    else:
        resultado = n * fatorial(n - 1, nivel + 1)
        print(f"Fatorial de {n} é {resultado}")
        return resultado

# Entrada do usuário
numero = int(input())
fatorial(numero)

```
